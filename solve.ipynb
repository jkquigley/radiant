```python
##############################
# Solves -u'' + u = f        #
# With zero Neumann boundary #
##############################
import radiant as rad
import numpy as np


# Problem Parameters
a = 0
b = 2 * np.pi


def u(x):
    return np.cos(x * 2 * np.pi / (b - a))


def f(x):
    return ((2 * np.pi / (b - a)) ** 2 + 1) * np.cos(x * 2 * np.pi / (b - a))
```

```python
# Parameters
d = 1
k = 1
delta = 1.9
N = 50

# Solve for approximate solution
u_approx, A, fs, centres, alphas = rad.solve(a, b, f, N, d, k, delta, precond=False)
error = rad.error(u, u_approx, a, b)
cond = np.linalg.cond(A)

print("N:", N)
print("L2 Relative Error:", error)
print("System condition number:", cond)
```
